```python
import os
import numpy as np 
```

```python
distance_file = os.path.join('distance_data_headers.csv')
distances = np.genfromtxt(fname = distance_file, delimiter = ',', dtype = 'unicode')
print(distances)
```

```python
headers = distances[0]
data = distances[1:]
data = data.astype(np.float)
```

```python
print(headers)
print(data)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure()
plt.plot(data[:,0],data[:,1]) # x's are every row just the 0th col, and y's every row just 1st col looks exactly the same as data[;,1] since pyplot automatically numbers the x if not specified 
```

```python
# labels 
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstroms)')
fig_1 = plt.plot(data[:,0], data[:,1], label = sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
# plot multiple in one figure

plt.figure()
for col in range(1, len(headers)):   # for plotting each col in one fig.
    fig_1 = plt.plot(data[:,0], data[:,col], label=headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstroms)')
#plt.plot(data[:,0], data[:,1], label = headers[1]) # how to pick col 1 & 2 to fig. .. 1st example. 
#plt.plot(data[:,0], data[:,2], label = headers[2])

plt.legend()

```

```python
# print each function as seperate files 

colors = ['r', 'g', 'b', 'c', 'm']

for col in range(1, len(headers)):   # for plotting each col in one fig.
    fig_1 = plt.plot(data[:,0], data[:,col], label=headers[col], color=colors[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (Angstroms)')
    plt.legend()
    plt.savefig('all_samples_330dpi.png', dpi=300)
    plt.figure()
```
